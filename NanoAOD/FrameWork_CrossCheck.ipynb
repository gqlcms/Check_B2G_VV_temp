```python
def eval_expr(ak_array, expr, mask=None):
    """A function that can do `eval` to the awkward array, immitating the behavior of `eval` in pandas."""
    
    def get_variable_names(expr, exclude=['awkward', 'ak', 'np', 'numpy', 'math']):
        """Extract variables in the expr"""
        import ast
        root = ast.parse(expr)
        return sorted({node.id for node in ast.walk(root) if isinstance(node, ast.Name) and not node.id.startswith('_')} - set(exclude))

    tmp = {k:ak_array[k] if mask is None else ak_array[k].mask[mask] for k in get_variable_names(expr)}
    tmp.update({'math': math, 'numpy': np, 'np': np, 'awkward': ak, 'ak': ak})
#     print('eval expr: ', expr, '\nvars', get_variable_names(expr))
    return eval(expr, tmp)

class NanoAODFrameWork_CrossCheck():
    
    def __init__(self, ConfigFile ):
        
        self.Branch = {}
        self.Ntuple = {}
        self.Array = {}
        self.Arrow = {}
        self.DataFrame = {}
        self.RDataFrame = {}
        self.Selection = {}
        self.Preselection = {}
        
        exec(open(ConfigFile,'r').read())
        
        
    def MiniAODNTuple_Branch(self, **kwargs ):
        FileLabel = kwargs.get('FileLabel', self.filelabel['miniAOD'])
        events = uproot.open(self.miniAOD['filename'][FileLabel])['treeDumper/EDBRCandidates']
        dic_ = events.typenames()
        list_ = dic_.keys()
        self.Branch['miniAOD'] = [list_,dic_]
        return list_, dic_
    
    
    def NanoAODNTuple_Branch(self, **kwargs ):
        FileLabel = kwargs.get('FileLabel', self.filelabel['nanoAOD'])
        events = uproot.open(self.nanoAOD['filename'][FileLabel])['t']
        dic_ = events.typenames()
        list_ = dic_.keys()
        self.Branch['nanoAOD'] = [list_,dic_]
        return list_, dic_
    
    
    def Compare_MiniAOD_NanoAOD_Branch(self):
        if len(self.Branch['miniAOD'][0]) == 0 : self.MiniAODNTuple_Branch() 
        if len(self.Branch['nanoAOD'][0]) == 0 : self.NanoAODNTuple_Branch() 
        self.Branch['comparison'] = {}
        miniAOD = set(self.Branch['miniAOD'][0]) ; nanoAOD = set(self.Branch['nanoAOD'][0]) ; common = miniAOD & nanoAOD;
        self.Branch['comparison']['common'] = list(miniAOD & nanoAOD) ;
        self.Branch['comparison']['miniAODOnly'] = list(miniAOD - common) ;
        self.Branch['comparison']['nanoAODOnly'] = list(nanoAOD - nanoAOD) ;
        return self.Branch['comparison']
    
    
    def GetNTuple(self,  **kwargs):
        miniAOD_FileLabel = kwargs.get('miniAOD_FileLabel', self.filelabel['miniAOD'])
        nanoAOD_FileLabel = kwargs.get('nanoAOD_FileLabel', self.filelabel['nanoAOD'])
        self.Ntuple['miniAOD'] = uproot.open(self.miniAOD['filename'][miniAOD_FileLabel])['treeDumper/EDBRCandidates']
        self.Ntuple['nanoAOD'] = uproot.open(self.nanoAOD['filename'][nanoAOD_FileLabel])['t']
    
    
    def NtupleToArray(self, **kwargs):
        if len(self.Branch) == 0 : self.MiniAODNTuple_Branch(); self.NanoAODNTuple_Branch();
        if len(self.Branch['miniAOD'][0]) == 0 : self.MiniAODNTuple_Branch() 
        if len(self.Branch['nanoAOD'][0]) == 0 : self.NanoAODNTuple_Branch() 
        miniAOD_branch = kwargs.get('miniAOD_branch', self.Selection['necessary variable'].get( 'miniAOD' , self.Branch['miniAOD'][0] ) )
        nanoAOD_branch = kwargs.get('nanoAOD_branch', self.Selection['necessary variable'].get( 'nanoAOD' , self.Branch['nanoAOD'][0] ) )
        if len(self.Ntuple) ==0 :
            self.GetNTuple()
            
        self.Array['miniAOD'] = self.Ntuple['miniAOD'].arrays(miniAOD_branch, library="ak")
        # be careful : NanoAOD hold p4, can not transfer to dataframe
        # this filter_name="/([A-Za-Z_]*)p4/i"doesn't work
        # if the vector is inside the dataframe, than this will return Ntuple not dataframe
        # https://uproot.readthedocs.io/en/latest/basic.html#nested-data-structures
#         nanoAOD_branch = [i for i in nanoAOD_branch if 'p4' not in i]
        self.Array['nanoAOD'] = self.Ntuple['nanoAOD'].arrays(nanoAOD_branch, library="ak")
#         self.DataFrame['nanoAOD'] = self.Ntuple['nanoAOD'].arrays(nanoAOD_branch, library="pd", how="outer")
    
    
    def Compare_MiniAOD_NanoAOD_PreSelection(self):
        if len(self.Array) ==0 :
            self.NtupleToArray()
        miniAOD = ak.to_list(self.Array['miniAOD'][self.Selection['necessary variable']['miniAOD_eventNumber']])
        nanoAOD = ak.to_list((self.Array['nanoAOD'][self.Array['nanoAOD'][self.Preselection['Preselection_cut']] > 0])[self.Selection['necessary variable']['nanoAOD_eventNumber']])
        miniAOD_set = set(list(miniAOD))
        nanoAOD_set = set(list(nanoAOD))
        Common = miniAOD_set & nanoAOD_set ; miniAOD_only = miniAOD_set - Common ;  nanoAOD_only = nanoAOD_set - Common ;  
        self.Preselection['common'] = Common ; self.Preselection['miniAOD_only'] = miniAOD_only ; self.Preselection['nanoAOD_only'] = nanoAOD_only ; 
        return Common, miniAOD_only, nanoAOD_only
        
        
    def Ak_to_Pa_to_Pandas(self):
        # nanoAOD to pandas
        # change the vector to array
        for i in self.vector['nanoAOD']:
            if i in self.Selection['necessary variable']['nanoAOD']:
                self.Array['nanoAOD'][i] = ak.pad_none(self.Array['nanoAOD'][i], axis = 1, target = self.vector['nanoAOD'][i], clip=True)
        
        for i in ['miniAOD', 'nanoAOD']:
            self.Arrow[i] = ak.to_arrow_table(self.Array[i])
            self.DataFrame[i] = (self.Arrow[i]).to_pandas()
            
    def Select_DiffEvent_AkwardArray(self):
        # this function is used to display the different events between the miniAOD and NanoAOD
        if sum([len(self.Preselection['event_list'][i]) for i in self.Preselection['event_list']]) == 0:
            self.Compare_MiniAOD_NanoAOD_PreSelection()
        
        # _cut1 is used to select events which only miniAOD or nanoAOD have
        for i in ['miniAOD', 'nanoAOD']:
            self.DataFrame[i+'_cut1'] = self.DataFrame[i][self.DataFrame[i][self.Selection['necessary variable'][i+'_eventNumber']].map(lambda x: x in self.Preselection[i+'_only'])]
        
        # nanoAOD_cut2 is used for the events pass miniAOD selection that not passed nanoAOD selection
        self.DataFrame['nanoAOD_cut2'] = self.DataFrame['nanoAOD'][self.DataFrame['nanoAOD'][self.Selection['necessary variable'][i+'_eventNumber']].map(lambda x: x in self.Preselection['miniAOD_only'])]
        self.DataFrame['nanoAOD_cut3'] = self.DataFrame['nanoAOD'][self.DataFrame['nanoAOD'][self.Selection['necessary variable'][i+'_eventNumber']].map(lambda x: x in self.Preselection['nanoAOD_only'])]
        
    
    def Pandas_Array_to_Mutibranch(self):
        # change the array to mutiple branch for display
        events_list = {
            'miniAOD_cut1' : 'miniAOD',
            'nanoAOD_cut1' : 'nanoAOD',
            'nanoAOD_cut2' : 'nanoAOD',
        }
        for i in events_list.keys():
            for vector_branch in self.vector[events_list[i]]:
                for j in range(len(self.DataFrame[i][vector_branch].iloc[0])):
                    self.DataFrame[i][vector_branch+'_'+str(j)] = self.DataFrame[i][vector_branch].map(lambda x: x[j])
                
                
    def Pandas_Array_to_Mutibranch_OneFrame(self, frame, vector_list):
        # change the array to mutiple branch for display
        for vector_branch in vector_list:
            for j in range(len(self.DataFrame[frame][vector_branch].iloc[0])):
                self.DataFrame[frame][vector_branch+'_'+str(j)] = self.DataFrame[frame][vector_branch].map(lambda x: x[j])

                
    def Pandas_LenOfArray(self, nanoAOD = 'nanoAOD_cut2'):
        # get the array's lenth
        # only used for nanoAOD
        for vector_branch in self.vector_len:
            self.DataFrame[nanoAOD][vector_branch+'_num'] = self.DataFrame[nanoAOD][vector_branch].map(lambda x: len(x))
    
    def Display_OneFrame_Diffevents(self, frame = 'nanoAOD', branch_label = 'nanoAOD', Max_nevents = 10):
        display((self.DataFrame[frame][self.DisplayBranch[branch_label]][:Max_nevents]).T)
        
    def Pd_WithCut(self, frame, cut_name):
        cut = eval_expr(self.DataFrame[frame], self.cuts[cut_name] )
        self.DataFrame[frame+cut_name] = self.DataFrame[frame][cut]
        
        
    def Display_TwoFrame(self, nanoAOD = 'nanoAOD_cut2', miniAOD = 'miniAOD_cut1',  Max_nevents = 10):
        events_list = list(set(self.DataFrame[nanoAOD][self.Selection['necessary variable']['nanoAOD_eventNumber']] ) & set(self.DataFrame[miniAOD][self.Selection['necessary variable']['miniAOD_eventNumber']] ))
        frame_label = {
            miniAOD: 'miniAOD',
            nanoAOD: 'nanoAOD',
        }
        print('\033[1;34;40m '+'number of events pass miniAOD Ntuple cut and not pass nanoAOD Ntuple cut : '+str(len(events_list))+'\033[0m')
        print('\033[1;34;40m '+'number of events pass nanoAOD Ntuple cut and not pass miniAOD Ntuple cut : '+str(len(self.Preselection['nanoAOD_only']))+'\033[0m')
        print('\033[1;34;40m '+'number of events pass miniAOD both nanoAOD Ntuple cut : '+str(len(self.Preselection['common']))+'\033[0m')
        
        for i in events_list[:Max_nevents]:
            combine = []
            for j in [miniAOD, nanoAOD]:
                exec(j+'_f = self.DataFrame[j][self.DataFrame[j][self.Selection["necessary variable"][frame_label[j]+"_eventNumber"]] == i]')
                exec(j+'_f = '+j+'_f[self.DisplayBranch[frame_label[j]]]')
                eval(j+'_f').index = [index for index in ['miniAOD', 'nanoAOD'] if index in j]
                combine.append(eval(j+'_f'))
            combine_df = pd.concat(combine)
            df_columns = combine_df.columns.tolist()
            df_columns_new = []
            for i in self.comparison:
                Comparison_in_Branch = True
                for j in self.comparison[i]:
                    Comparison_in_Branch = Comparison_in_Branch & (j in df_columns)
                if Comparison_in_Branch:
                    for j in self.comparison[i]:
                        df_columns.remove(j)
                        df_columns_new.append(j)
            df_columns_new += df_columns
            combine_df = combine_df[df_columns_new]
            display(combine_df.T)
            
    def Prepare_Rdataframe(self, **kwargs):
        framelabel = kwargs.get('framelabel', 'miniAOD')
        label = kwargs.get('label', self.filelabel[framelabel]) # www, ttbar
        treeName = kwargs.get('treeName', self.treeName[framelabel])
        fileName = kwargs.get('fileName', self.fileName[framelabel][label])
        RDataFrame_Name = kwargs.get('RDataFrame_Name', 'temp')
        self.RDataFrame[RDataFrame_Name] = ROOT.RDataFrame(treeName, fileName)
    
    def Rdataframe_Pandas(self, **kwargs):
        RDataframe_Name = kwargs.get('RDataframe_Name', 'temp')        
        PandasDataframe_Name = kwargs.get('PandasDataframe_Name', 'temp')
        framelabel = kwargs.get('framelabel', 'miniAOD')
        columns = kwargs.get('columns', self.Rdataframe_to_Pandas_columns[framelabel])
        rdf = self.RDataFrame[RDataframe_Name]
        # in this step, the RDataframe can not contain vector branch
        pddf = pd.DataFrame(rdf.AsNumpy(columns=columns))
        self.DataFrame[PandasDataframe_Name] = pddf
        
    def Rdataframe_addcolumns(self, **kwargs):
        RDataframe_Name = kwargs.get('RDataframe_Name', 'temp')        
        framelabel = kwargs.get('framelabel', 'miniAOD')
        columns_dic = kwargs.get('columns_dic', self.Rdataframe_newcolumns[framelabel])
        columns_list = kwargs.get('columns_list', self.Rdataframe_newcolumns['column_order'][framelabel])
        rdf = self.RDataFrame[RDataframe_Name]
        for i in columns_list:
            rdf = rdf.Define(i, columns_dic[i])
        self.RDataFrame[RDataframe_Name] = rdf
        
    def Pandas_DataFrame_Sets(self, **kwargs):
        # produce intersection, difference set of two columns from two pandas frame
        Dataframe_Name1 = kwargs.get('Dataframe_Name1', 'temp1')
        Dataframe_Name2 = kwargs.get('Dataframe_Name2', 'temp2')
        column1 = kwargs.get('column1', 'temp1')
        column2 = kwargs.get('column2', 'temp1')
        dfc1 = set(self.DataFrame[Dataframe_Name1][column1])
        dfc2 = set(self.DataFrame[Dataframe_Name2][column2])
        common = dfc1 & dfc2
        df1only = dfc1 - common
        df2only = dfc2 - common
        print('number of common : ',len(common))
        print('number of ' , Dataframe_Name1 , ' only : ',len(df1only), '%.2f' %(float(len(df1only))/float(len(common)+len(df1only))) )
        print('number of ' , Dataframe_Name2, ' only : ', len(df2only), '%.2f' %(float(len(df2only))/float(len(common)+len(df2only))) )
        return common, df1only, df2only
    
    def PandasFrame_Selection(self, **kwargs):
        # the input for cut can be list or cut name
        # if the input is cut name: 
        # 1. all the variable used cuts are int, if > 0, pass the cut
        # if the input is list:
        # the use this list to select pandas dataframe's raws
        InDataframe_Name = kwargs.get('InDataframe_Name', 'temp1')
        OutDataframe_Name = kwargs.get('OutDataframe_Name', 'temp2')
        cut_name = kwargs.get('cut_name', [])
        event_list = kwargs.get('event_list', [])
        event_column1 = kwargs.get('event_column1', 'temp1')
        df = self.DataFrame[InDataframe_Name]
        if len(cut_name) > 0:
            for cut in cut_name:
                df = df[df[cut] > 0]
        if len(event_list) > 0:
            df = df.loc[df[event_column1].isin(event_list)]
        
        self.DataFrame[OutDataframe_Name] = df
         
    def Display_Pandas_DataFrame(self, **kwargs):
        Max_nevents = kwargs.get('Max_nevents', 10)
        events_list = kwargs.get('events_list', [])
        Dataframe_Name1 = kwargs.get('Dataframe_Name1', 'temp1')
        Dataframe_index1 = kwargs.get('Dataframe_index1', ['index1'])
        Dataframe_Name2 = kwargs.get('Dataframe_Name2', 'temp2')
        Dataframe_index2 = kwargs.get('Dataframe_index2', ['index2'])
        event_column1 = kwargs.get('event_column1', 'temp1')
        event_column2 = kwargs.get('event_column2', 'temp2')
        df_columns_new = kwargs.get('df_columns_new', ['temp2'])
        
        df1 = self.DataFrame[Dataframe_Name1]
        df2 = self.DataFrame[Dataframe_Name2]
        
        for i in events_list[:Max_nevents]:
            df1c = df1[df1[event_column1] == i] ; df1c.index = Dataframe_index1
            df2c = df2[df2[event_column2] == i] ; df2c.index = Dataframe_index2
            combine = [df1c, df2c]
            combine_df = pd.concat(combine)
            df_columns = combine_df.columns.tolist()
            combine_df = combine_df[df_columns_new]
            display(combine_df.T)
            
    def Plots_Variables(self, **kwargs):
        Comparison = kwargs.get('Comparison', False) # comparison two dataframe or plot single dataframe
        Dataframe_Name1 = kwargs.get('Dataframe_Name1', 'temp1')
        Dataframe_Name2 = kwargs.get('Dataframe_Name2', 'temp2')
        varlist = kwargs.get('varlist', ['temp1'])
        Subplots_Split = kwargs.get('Subplots_Split', (int(len(varlist)**0.5),int(len(varlist)**0.5)+1))
        df1 = self.DataFrame[Dataframe_Name1]
        df2 = self.DataFrame[Dataframe_Name2]
        if Comparison:
            plt.figure(figsize=(5*Subplots_Split[0],5*Subplots_Split[1]), dpi=100)
            for i,var in enumerate(var_list):
                plt.subplot(Subplots_Split[0],Subplots_Split[1],i+1)
                plt.hist(df1[var[0]['name']],  bins=var[0]['bin'][0], range=(var[0]['bin'][1], var[0]['bin'][2]),color = 'Red', label = '1', histtype = 'step')
                plt.hist(df2[var[1]['name']],  bins=var[1]['bin'][0], range=(var[1]['bin'][1], var[1]['bin'][2]),color = 'Blue', label = '2', histtype = 'step')
                plt.title(var[0]['tile'])
                plt.legend(loc=9)
                
    def Sort_Pandasframe(self, **kwargs):
        Dataframe_Name1 = kwargs.get('Dataframe_Name', 'temp1')
        sort_column1 = kwargs.get('sort_column', 'temp1')
        self.DataFrame[Dataframe_Name1] = self.DataFrame[Dataframe_Name1].sort_values(by=[sort_column1])
        
    def Merge_Twodataframe(self, **kwargs):
        InDataframe_Name1 = kwargs.get('InDataframe_Name1', 'temp1')
        InDataframe_Name2 = kwargs.get('InDataframe_Name2', 'temp1')
        OutDataframe_Name = kwargs.get('OutDataframe_Name', 'temp2')
        columns_dic = kwargs.get('columns_dic', 'temp2')
        # change the columns name 
        df1 = self.DataFrame[Dataframe_Name1]
        df2 = self.DataFrame[Dataframe_Name2]
        df1 = df1.reset_index()
        df2 = df2.reset_index()
        merge_columns = []
        for i in columns_dic:
            df1 = df1.rename(columns=columns_dic[i][0])
            df2 = df2.rename(columns=columns_dic[i][1])
            newcolumn_name1 = columns_dic[i][0][list(columns_dic[i][0].keys())[0]]
            newcolumn_name2 = columns_dic[i][1][list(columns_dic[i][1].keys())[0]]
            merge_columns.append(pd.Series(df1[newcolumn_name1], name=newcolumn_name1))
            merge_columns.append(pd.Series(df2[newcolumn_name2], name=newcolumn_name2))
        # merge two dataframe
        self.DataFrame[OutDataframe_Name] = pd.concat(merge_columns, axis=1)
        
    def Compare_Variables_In_miniAOD_and_NanoAOD(self, **kwargs):
        Dataframe_Name = kwargs.get('Dataframe_Name', 'temp1')
        variables = kwargs.get('variables', 'temp1')
        
        df = self.DataFrame[Dataframe_Name]
        for i in variables:
            column1 = variables[i]['column1']
            column2 = variables[i]['column2']
            event_column1 = variables[i]['event_column1']
            event_column2 = variables[i]['event_column2']
            Allow_Difference = variables[i]['Allow_Difference']
            
            diff = (df[column1] - df[column2]).abs()/(df[column1].abs())
            common_df = df[diff<Allow_Difference]
            diff_df = df[diff>=Allow_Difference]
            common_events = common_df[event_column1]
            diff_events = diff_df[event_column1]
            
            print('common:',len(common_events),'diff:',len(diff_events), )
            print(column1,':', ' %.2f '%(float(len(diff_events)*100)/(float(len(diff_events))+float(len(common_events))))+'%')
            
        if len(variables) == 1:
            return common_events, diff_events
    
    
class variables_comparison_config():
    
    def __init__(self, ConfigFile ):
        
        exec(open(ConfigFile,'r').read())
```
