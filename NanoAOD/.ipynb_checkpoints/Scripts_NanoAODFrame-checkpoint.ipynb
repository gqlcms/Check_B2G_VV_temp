```python
def NanoAOD_Content(TreeKey):
    KeyList = []
    KeyDic = {}
    for i in TreeKey:
        first_part = str(i).split("_")[0]
        if first_part not in KeyList:
            KeyList.append(first_part)
    for i in KeyList:
        KeyDic[i] = []
    for i in TreeKey:
        first_part = str(i).split("_")[0]
        full_part = str(i)
        KeyDic[first_part].append(full_part)
    return KeyList,KeyDic

def Branch_Group(SaveBranch,KeyDic):
    Branch = {}
    for i in SaveBranch:
        Branch[i] = KeyDic[i]
    return Branch

def GetAllTyep(AllType):
    TypeList = []
    for i in AllType:
        if AllType[i] not in TypeList:
            TypeList.append(AllType[i])
    return TypeList

def Branch_Type(BranchType):
    Vector = False ; 
    if '[]' in BranchType:
        Vector = True
    BranchTypeDic = {'int':'int','bool':'int','float':'float'}
    for i in BranchTypeDic:
        if i in BranchType:
            Type = BranchTypeDic[i]
    return (Vector,Type)

def BranchCode(BranchName,BranchType,prefix,LoopLabel = 'i'):
    CreateBranch_Templete ={
        'single' : 'ana.tx.createBranch<{0}>                  ("{1}");',\
        'vector' : 'ana.tx.createBranch<vector<{0}>>                  ("{1}");',\
    }
    FillBranch_Templete = {
        'vector':'ana.tx.pushbackToBranch<{0}>("{1}", nt.{2}()[{3}]);',\
        'single':'ana.tx.setBranch<{0}>("{1}", nt.{2});',\
    }
    BranchType_ = Branch_Type(BranchType)
    if BranchType_[0]:
        Create_Str = CreateBranch_Templete['vector'].format(BranchType_[1],prefix+"_"+BranchName)
        Fill_Str = FillBranch_Templete['vector'].format(BranchType_[1],prefix+"_"+BranchName,BranchName, LoopLabel)
    else:
        Create_Str = CreateBranch_Templete['single'].format(BranchType_[1],prefix+"_"+BranchName)
        Fill_Str = FillBranch_Templete['single'].format(BranchType_[1],prefix+"_"+BranchName,BranchName)
    return Create_Str,Fill_Str

def BranchCodeAll(Branch_Dic, Type_Dic,prefix='Lepton1',Write = False, CodeText = 'code.txt', LoopLabel='i'):
    CreateList = []; FillList = [];
    for i in Branch_Dic:
        for j in Branch_Dic[i]:
            Create_Str,Fill_Str = BranchCode(j,Type_Dic[j],prefix, LoopLabel)
            CreateList.append(Create_Str)
            FillList.append(Fill_Str)
    if Write:
        Write_str = ''
        for i in CreateList:
            Write_str += i+'\n'
        Write_str += '\n\n\n'
        for i in FillList:
            Write_str += i+'\n'
        with open(CodeText,'w') as f:
            f.write(Write_str)
    return CreateList, FillList

class NanoCode():
    
    import re
    
    def __init__(self ):
        self.CreateBranch_Templete ={
            'single' : 'ana.tx.createBranch<{0}>                  ("{1}");',\
            'vector' : 'ana.tx.createBranch<vector<{0}>>                  ("{1}");',\
        }
        self.FillBranch_Templete = {
            'vector':'ana.tx.pushbackToBranch<{0}>("{1}", nt.{2}()[{3}]);',\
            'single':'ana.tx.setBranch<{0}>("{1}", nt.{2});',\
        }
        self.Variable_Matching_Gen = {
            'pt' : 'GenPart_pt[]',
            'eta' : 'GenPart_eta[]',
            'phi' : 'GenPart_phi[]',
            'mass' : 'GenPart_p4()[].mass()',
            'pdg' : 'GenPart_pdgId()[]',
            'energy' : 'nt.GenPart_p4()[].energy()', 
        }
        self.CreateBranch = {
            'ana.tx.setBranch': (re.compile(r'ana.tx.setBranch<(\w*)>\((\s*)"(\w*)"') , 'ana.tx.createBranch<{0}>                  ("{1}");'),
            'ana.tx.pushbackToBranch': (re.compile(r'ana.tx.pushbackToBranch<(\w*)>\((\s*)"(\w*)"') , 'ana.tx.createBranch<vector<{0}>>                  ("{1}");'),
        }
    
    
    def CreateBranch_From_MiniAODCode(self, MiniCode, VariableType = 'float', prefix = '', LoopLabel = 'i', Print = True):
        
        MiniCode = MiniCode.split('\n')
        NanoCode = ''
        for i in MiniCode:
            VariableName = i.split('=')[0].replace(" ","")
            vector = False
            if ( '[' in  VariableName ) & ( ']' in  VariableName ) : vector = True
            if vector:
                Fill_Str = self.FillBranch_Templete['vector'].format(VariableType,prefix+VariableName,VariableName, LoopLabel)
            else:
                Fill_Str = self.FillBranch_Templete['single'].format(VariableType,prefix+VariableName,'')
            NanoCode += Fill_Str+'\n'
        if Print:
            print(NanoCode)
        return NanoCode
    
    def CreateBranch_From_MiniAODCode_Gen(self, MiniCode, VariableType = 'float', prefix = '', LoopLabel = '', Print = True):
        
        MiniCode = MiniCode.split('\n')
        NanoCode = ''
        for i in MiniCode:
            VariableName = i.split('=')[0].replace(" ","")
            vector = False
            if ( '[' in  VariableName ) & ( ']' in  VariableName ) : vector = True
            NanoBranch = '' 
            for key in self.Variable_Matching_Gen.keys(): 
                if key in VariableName:
                    NanoBranch = self.Variable_Matching_Gen[key]
            if vector:
                VariableName = VariableName.split("[")[0]
                Fill_Str = self.FillBranch_Templete['vector'].format(VariableType,prefix+VariableName,NanoBranch, LoopLabel)
            else:
                Fill_Str = self.FillBranch_Templete['single'].format(VariableType,prefix+VariableName,NanoBranch)
            
            NanoCode += Fill_Str+'\n'
        if Print:
            print(NanoCode)
        return NanoCode
    
    def NanoAOD_CreateBranch(self, Fill):
        FillString_out = ''
        FillString = [j for j in Fill.split("\n") if not re.match(r'(\s*)//(\s*)',j) ]
        BranchList = []
        for i in FillString:
            for j in self.CreateBranch:
                if j in i:
                    Matched = ((self.CreateBranch[j][0]).search(i)).groups()
                    if Matched[2] not in BranchList:
                        BranchList.append(Matched[2])
                        FillString_Temp = self.CreateBranch[j][1].format(Matched[0],Matched[2])
                        FillString_out += FillString_Temp+"\n"
        return FillString_out
```

```python
NanoCode_I1 = NanoCode()
with open('create.txt','w') as f:
    f.write(NanoCode_I1.NanoAOD_CreateBranch(open('fill.txt','r').read()))
```
